```python
dados = pd.read_csv("dados_vendas.csv")
```

```python
dados
```

```python
#media dos valores das vendas
media_vendas = dados["Valor da venda"].median()
print(media_vendas)
```

```python
#mediana dos valores de venda
mediana_vendas = dados["Valor da venda"].quantile(0.5)
print(mediana_vendas)
```

```python
#valor minimo de vendas
minimo_vendas = dados["Valor da venda"].min()
print(minimo_vendas)
```

```python
#valor maximo de vendas
maximo_vendas = dados["Valor da venda"].max()
print(maximo_vendas)
```

```python
#valor da variancia
variancia_vendas = dados["Valor da venda"].var()
print(variancia_vendas)
```

```python
#desvio padrao de vendas
desvio_padrao_vendas = dados["Valor da venda"].std()
print(desvio_padrao_vendas)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(dados["Valor da venda"], bins=20, edgecolor="black")
plt.xlabel("Valor da venda")
plt.ylabel("Frequência")
plt.title("Histograma de Vendas")
plt.show
```

```python
df['Data da venda'] = pd.to_datetime(df['Data da venda'])

# Criando o gráfico de linha para visualizar os valores das vendas ao longo das datas
plt.figure(figsize=(12, 6))
plt.plot(df['Data da venda'], df['Valor da venda'], marker='o', linestyle='-', color='b')
plt.title('Valores das Vendas ao Longo das Datas')
plt.xlabel('Data da Venda')
plt.ylabel('Valor da Venda')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

```python
dados['Quantidade vendida'].plot(kind='line', figsize=(8, 4), title='Quantidade vendida')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
plt.figure(figsize=(10, 6))
df['Produto vendido'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição dos Produtos Vendidos')
plt.xlabel('Produto Vendido')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Gráfico de pizza para "Método de pagamento"
plt.figure(figsize=(10, 6))
df['Método de pagamento'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'green', 'red'])
plt.title('Distribuição dos Métodos de Pagamento')
plt.ylabel('')  # Remove o label do eixo y para melhorar a aparência
plt.show()
```

```python
# Gráfico de barras para "Região de venda"
plt.figure(figsize=(10, 6))
df['Região de venda'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição das Vendas por Região')
plt.xlabel('Região')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
