```python
df = pd.DataFrame(dados_vendas)
```

Calculando as métricas solicitadas 

```python
media = df['Valor da venda'].mean()
mediana = df['Valor da venda'].median()
minimo = df['Valor da venda'].min()
maximo = df['Valor da venda'].max()
desvio_padrao = df['Valor da venda'].std()
```

```python
print(f'A média de vendas foi {media}')
print(f'A mediana foi {mediana}')
print(f'A venda mínima foi {minimo}')
print(f'A venda máxima foi {maximo}')
print(f'O desvio padrão foi {desvio_padrao}')
```
