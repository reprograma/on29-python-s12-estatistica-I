```python
colunas = ["Data da venda", "Valor da venda", "Produto vendido", "Quantidade vendida", "Método de pagamento", "Região de venda"]
arquivo = "dados_vendas.csv"

with open('./dados_vendas.csv', mode='w', newline='', encoding='utf-8') as csvfile:
    escritor = csv.DictWriter(csvfile, delimiter=",", fieldnames=colunas)
    escritor.writeheader()
    escritor.writerows(dados_vendas)
```

```python
df = pd.read_csv('dados_vendas.csv', sep=",")
df
```

1.  Calcular a média

```python
df_total_vendas = df['Valor da venda'].sum()
print(f"Valor total vendas: R${df_total_vendas}")
```

```python
media_vendas = df['Valor da venda'].mean()
print(f"Média das vendas: {media_vendas}")
```

2. Mediana

```python
mediana_vendas = df["Valor da venda"].median()
```

3. mínimo
4. máximo

```python
df_vendas_min = df['Valor da venda'].min()
print(f"Venda mínima de: R${df_vendas_min}")
```

```python
df_vendas_max = df['Valor da venda'].max()
print(f"Venda máxima de R${df_vendas_max}")
```

5. desvio padrão dos valores das vendas.

```python
df_desvio_vendas = df['Valor da venda'].std()
print(f"O desvio padrão das vendas é de R${df_desvio_vendas:.2f}")
```

```python
print(f"1. total de vendas: R${df_total_vendas:.2f}\n2. Média das vendas: R${media_vendas:.2f}\n3. Mediana das vendas: R${mediana_vendas:.2f}\n4. Venda mínima: R${df_vendas_min:.2f}\n5. Venda máxima: R${df_vendas_max:.2f}\n6. Desvio Padrão dos valores: R${df_desvio_vendas:.2f}")
```

```python
# método de pagamento mais utilizado
receita_por_pagamento = df.groupby('Método de pagamento')['Valor da venda'].sum()

receita_por_pagamento.plot(kind = 'line', marker = 'o', label='Total')

plt.title("Metodo de pagamento")
plt.xlabel('Tipo de transação')
plt.ylabel('Receita')
plt.legend()

for i, (index, value) in enumerate(receita_por_pagamento.items()):
    plt.text(i, value, str(value), ha='center', va='bottom')
```

```python
# Produtos mais Vendidos
top_produtos = df.groupby('Produto vendido')['Quantidade vendida'].sum().head(6).sort_values(ascending=False)

top_produtos.plot(kind= 'bar', edgecolor='green', color='skyblue')
plt.title('Produto Mais vendido')
plt.xlabel('Produto')
plt.ylabel('Quantidade')

for i, (index, value) in enumerate(top_produtos.items()):
    plt.text(i, value, str(value), ha='center', va='bottom')
```

```python
# média de vendas diaria
df["Data da venda"] = pd.to_datetime(df['Data da venda'])

receita_diaria = df.groupby(df["Data da venda"].dt.strftime('%d'))['Valor da venda'].sum()

receita_diaria.plot(kind = 'line', marker = 'o', label='Total')

plt.title("Receita Diária")
plt.xlabel('Data')
plt.ylabel('Receita')
plt.legend()

for i, (index, value) in enumerate(receita_diaria.items()):
    plt.text(i, value, str(value), ha='center', va='bottom')
```
