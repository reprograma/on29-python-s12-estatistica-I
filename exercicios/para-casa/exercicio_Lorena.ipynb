```python
df = pd.DataFrame(dados_vendas)
display(df)
```

```python
#Criando um array somente da coluna 'Valor da venda' para cálculo das medidas estatísticas
valores_vendas = df['Valor da venda'].values
display(valores_vendas)
```

```python
#Calculando as medidas estatísticas 
media = np.mean(valores_vendas)
mediana = np.median(valores_vendas)
minimo = np.min(valores_vendas)
maximo = np.max(valores_vendas)
desvio_padrao = np.std(valores_vendas)
```

```python
#Transformado os valores obtidos para R$
media_em_reais = f"R${media:_.2f}"
media_em_reais = media_em_reais.replace(".", ",").replace("_", ".")

mediana_em_reais = f"R${mediana:_.2f}"
mediana_em_reais = mediana_em_reais.replace(".", ",").replace("_", ".")

minimo_em_reais = f"R${minimo:_.2f}"
minimo_em_reais = minimo_em_reais.replace(".", ",").replace("_", ".")

maximo_em_reais = f"R${maximo:_.2f}"
maximo_em_reais = maximo_em_reais.replace(".", ",").replace("_", ".")

desvio_padrao_em_reais = f"R${desvio_padrao:_.2f}"
desvio_padrao_em_reais = desvio_padrao_em_reais.replace(".", ",").replace("_", ".")
```

```python
#Exibindo os resultados obtidos
print(f"A média do valor das vendas foi de {media_em_reais}.")
print(f"A mediana do valor das vendas é {mediana_em_reais}.")
print(f"O valor de venda mínimo foi de {minimo_em_reais}.")
print(f"O valor de venda máximo foi de {maximo_em_reais}.")
print(f"O desvio padrão do valor das vendas é de {desvio_padrao_em_reais}.")
```

```python
import matplotlib.pyplot as plt
```

```python
vendas_por_regiao = df.groupby('Região de venda')['Valor da venda'].sum()

vendas_por_regiao.plot(kind='bar', color='skyblue')
plt.title('Total de Vendas por Região')
plt.xlabel('Região')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
vendas_por_produto = df.groupby('Produto vendido')['Quantidade vendida'].sum()

vendas_por_produto.plot(kind='bar', color='skyblue')
plt.title('Total de Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
utilizacao_metodos_pagamento = df['Método de pagamento'].value_counts()

utilizacao_metodos_pagamento = utilizacao_metodos_pagamento.sort_values(ascending=False)

utilizacao_metodos_pagamento.plot(kind='bar', color='skyblue')
plt.title('Número de utilizações por método de pagamento')
plt.xlabel('Método de pagamento')
plt.ylabel('Número de utilizações')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```
