### Importando Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

### Criando o dataframe

```python
dados = pd.DataFrame(dados_vendas)
display(dados.head(5))
```

### Análise da Base de Dados

```python
# Tamanho da base

rows, columns = dados.shape

print(f"A base de dados tem {rows} linhas e {columns} colunas.")
```

```python
# Verificando os tipos de dados

dados.dtypes
```

```python
# Convertendo o tipo de dado do atributo "Data da venda"

dados['Data da venda'] = pd.to_datetime(dados['Data da venda'])

print(dados.dtypes)
```

```python
# Cálculo de medidas estatísticas da quantidade de produtos vendidos, utilizando numpy

from tabulate import tabulate

media_vendas = np.mean(dados['Quantidade vendida'])
mediana_vendas = np.median(dados['Quantidade vendida'])
maior_venda = np.max(dados['Quantidade vendida'])
menor_venda = np.min(dados['Quantidade vendida'])
desvpad_vendas = np.std(dados['Quantidade vendida'])

cards = [
    ["Média de quantidade de produtos vendidos na loja x", media_vendas],
    ["Mediana de quantidade de produtos vendidos na loja x", mediana_vendas],
    ["Maior quantidade de produtos vendidos na loja x", maior_venda],
    ["Menor quantidade de produtos vendidos na loja x", menor_venda],
    ["Valor do desvio padrão de quantidade de produtos vendidos na loja x", desvpad_vendas],
]

print("Os dados estatísticos relativos à quantidade de produtos vendidos são:\n")
print(tabulate(cards, tablefmt = "grid"))
```

```python
# Cálculo de medidas estatísticas do faturamento, utilizando numpy

from tabulate import tabulate

media_fat = round(np.mean(dados['Valor da venda']), 1)
mediana_fat = round(np.median(dados['Valor da venda']), 1)
maior_fat = round(np.max(dados['Valor da venda']), 1)
menor_fat = round(np.min(dados['Valor da venda']), 1)
desvpad_fat = round(np.std(dados['Valor da venda']), 1)

cards = [
    ["Média de faturamento", media_fat],
    ["Valor da Mediana relativa ao faturamento", mediana_fat],
    ["Maior faturamento", maior_fat],
    ["Menor faturamento", menor_fat],
    ["Valor do desvio padrão relativo ao faturamento", desvpad_fat],
]

print("Os dados estatísticos relativos ao faturamento são:\n")
print(tabulate(cards, tablefmt = "grid"))
```

```python
# distribuição da quantidade de produtos vendidos
sns.set(style="darkgrid")
plt.figure(figsize=(6, 4))
sns.histplot(dados['Quantidade vendida'], bins=5, kde=True, color='#6959CD', edgecolor='#D3D3D3')
plt.title('Distribuição da Quantidade de Produtos Vendidos', fontsize=16)
plt.xlabel('Quantidade de Produtos Vendidos', fontsize=14)
plt.ylabel('Frequência', fontsize=14)

counts, bin_edges = np.histogram(dados['Quantidade vendida'], bins=10)
for count, bin_edge in zip(counts, bin_edges):
    plt.text(bin_edge, count, str(count), fontsize=12, ha='center', va='bottom')
plt.show()
```

```python
# Total de Vendas por Produto

total_vendas_produto = dados.groupby('Produto vendido')['Valor da venda'].sum().sort_values(ascending=False)
total_vendas_produto.plot(kind='bar', title='Total de Vendas por Produto', xlabel='Produto', ylabel='Valor da venda', legend=False)
plt.show()
```

```python
# Análise de vendas por região
quantidade_vendida_regiao = dados.groupby('Região de venda')['Quantidade vendida'].sum().sort_values(ascending=False)

plt.fill_between(quantidade_vendida_regiao.index, quantidade_vendida_regiao.values, color ="#bdb76b", alpha=0.9)
plt.plot(quantidade_vendida_regiao.index, quantidade_vendida_regiao.values, color="Black", alpha=0.8, linewidth=5)

plt.title('Quantidade Vendida por Região')
plt.xlabel('Região')
plt.ylabel('Quantidade vendida')
plt.show()
```

```python
# Distribuição dos Métodos de Pagamento
dados['Método de pagamento'].value_counts().plot(kind='pie', title='Distribuição dos Métodos de Pagamento', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```
