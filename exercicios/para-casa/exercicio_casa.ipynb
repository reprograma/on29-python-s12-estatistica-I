```python
analise_regiao = df.groupby('Região de venda')['Valor da venda'].sum()
plt.bar(analise_regiao.index, analise_regiao.values)
plt.xlabel('Região da venda')
plt.ylabel('R$ venda')
plt.title('Valor da venda por Região')
plt.legend(['Valores de venda'], loc='upper left')
for i, valor in enumerate(analise_regiao.values):
    plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')
plt.show()
```

```python
histograma_vendas = df.groupby(['Produto vendido', 'Método de pagamento'])['Valor da venda'].sum().unstack(fill_value=0)
bar_width = 0.35
num_bars = len(histograma_vendas.index)
indices = np.arange(num_bars)
cores = ['red', 'green', 'violet']
for i, metodo in enumerate(histograma_vendas.columns):
    plt.bar(indices + i * bar_width, histograma_vendas[metodo], bar_width, color = cores[i], label=metodo)

plt.xlabel('Produto vendido')
plt.ylabel('R$ venda')
plt.title('Valor da venda por Produto e Método de pagamento')
plt.xticks(indices + bar_width/2,histograma_vendas.index)
plt.legend(title='Método de pagamento', loc='upper right', bbox_to_anchor=(1.15, 1.02))
for i, metodo in enumerate(histograma_vendas.columns):
    for j, valor in enumerate(histograma_vendas[metodo]):
        plt.text(indices[j]+i*bar_width, valor,f'{valor:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
